```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools_number = len(school_data_complete["school_name"].unique())
# Calculate the total number of students
total_students_number = len(school_data_complete["student_name"])
# Calculate the total budget
total_budget_amount = sum(school_data_complete["budget"].unique())
# Calculate the average math score
avrg_math_score = sum(school_data_complete["math_score"]) / total_students_number
# Calculate the average reading score
avrg_reading_score = sum(school_data_complete["reading_score"]) / total_students_number
# Calculate the overall passing rate (overall average score)
overall_average_score = (avrg_math_score + avrg_reading_score) / 2
# Calculate the percentage of students with a passing math score
pssng_math_score = school_data_complete.loc[school_data_complete['math_score'] >= 70 ,:]
# Calculate the percentage of students with a passing reading score
pssng_reading_score = school_data_complete.loc[school_data_complete['reading_score'] >= 70 ,:]
# Create a dataframe to hold the above results
passed_math = len(pssng_math_score["student_name"])
final_math = (passed_math / total_students_number) * 100
passed_reading = len(pssng_reading_score["student_name"])
final_reading = (passed_reading / total_students_number) * 100
# summary
summary1 = pd.DataFrame({"Total Schools": total_schools_number,     
                        "Total Students": total_students_number,
                        "Total Budget": total_budget_amount,
                         "Average Math Score": avrg_math_score,
                         "Average Reading Score": avrg_reading_score,
                         "%Passing Math": final_math,
                         "%Passing reading": final_reading,
                         "%Overall Passing Rate": [overall_average_score]})
summary1.style.format({"Total Budget":"${:,.2f}",
                        "%Passing Math":"%{:,.2f}",
                        "%Passing reading":"%{:,.2f}",
                         "%Overall Passing Rate":"%{:,.2f}" })



```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# School Name
school_group = school_data_complete.groupby(['school_name'])
# School Type
school_type = school_data.set_index(["school_name"])["type"]
# Total Students
total_students_number = school_group['student_name'].count()
# Total School Budget
total_school_budget = school_group['budget'].mean()
# Per Student Budget
per_student_budget =  school_group['budget'].mean() / school_group['student_name'].count()
# Average Math Score
avg_math = school_group['math_score'].mean()
# Average Reading Score
avg_reading = school_group['reading_score'].mean()
# % Passing Math
pssng_math_nm = pssng_math_score['school_name'].value_counts()
# % Passing Reading
pssng_reading_nm = pssng_reading_score['school_name'].value_counts()
# Overall Passing Rate (Average of the above two)
pssng_math_percent = (pssng_math_nm / total_students_number) * 100

pssng_reading_percent = (pssng_reading_nm / total_students_number) * 100

overall_pssng_avrg_percent = (pssng_math_percent + pssng_reading_percent) / 2
#-------------------------------------------------------------------------------------------------------------------#
# Create a dataframe to hold the above results

summary2 = pd.DataFrame({"School Type": school_type,     
                        "Total Students": total_students_number,
                        "Total School Budget": total_school_budget,
                        "Per Student Budget": per_student_budget,
                         "Average Math Score": avg_math,
                         "Average Reading Score": avg_reading,
                         "%Passing Math": pssng_math_percent,
                         "%Passing reading": pssng_reading_percent,
                         "%Overall Passing Rate": overall_pssng_avrg_percent})

summary2['Total School Budget'] = (summary2['Total School Budget']).apply(lambda x: '${:,.2f}'.format(x))
summary2['Per Student Budget'] = (summary2['Per Student Budget']).apply(lambda x: '{:,.2f}'.format(x))
summary2['%Passing Math'] = (summary2['%Passing Math']).apply(lambda x: '%{:,.2f}'.format(x))
summary2['%Passing reading'] = (summary2['%Passing reading']).apply(lambda x: '%{:,.2f}'.format(x))
summary2['%Overall Passing Rate'] = (summary2['%Overall Passing Rate']).apply(lambda x: '%{:,.2f}'.format(x))
summary2['Average Math Score'] = (summary2['Average Math Score']).apply(lambda x: '{:,.2f}'.format(x))
summary2['Average Reading Score'] = (summary2['Average Reading Score']).apply(lambda x: '{:,.2f}'.format(x))

top_five = summary2.sort_values(["%Overall Passing Rate"], ascending=False)


top_five.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a pandas series for each grade. Hint: use a conditional statement.

math_score_9th = school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby("school_name")["math_score"].mean()

math_score_10th = school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby("school_name")["math_score"].mean()

math_score_11th = school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby("school_name")["math_score"].mean()

math_score_12th = school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby("school_name")["math_score"].mean()  
 
# Group each series by school
    
grade_math_score={
    '9th':math_score_9th,
    '10th':math_score_10th,
    '11th':math_score_11th,
    '12th':math_score_12th,}

math_score_by_grade = pd.DataFrame(grade_math_score)
math_score_by_grade.index.name = None

math_score_by_grade['9th'] = (math_score_by_grade['9th']).apply(lambda x: '{:,.2f}'.format(x))
math_score_by_grade['10th'] = (math_score_by_grade['10th']).apply(lambda x: '{:,.2f}'.format(x))
math_score_by_grade['11th'] = (math_score_by_grade['11th']).apply(lambda x: '{:,.2f}'.format(x))
math_score_by_grade['12th'] = (math_score_by_grade['12th']).apply(lambda x: '{:,.2f}'.format(x))


math_score_by_grade.head(20)    
    
    
    
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
summary2["Average Math Score"] = pd.to_numeric(summary2["Average Math Score"])
summary2["Average Reading Score"] = pd.to_numeric(summary2["Average Reading Score"])

#summary2["Per Student Budget"] = summary2["Per Student Budget"].replace('$', '', regex=True).astype('float')
summary2["Per Student Budget"] = pd.to_numeric(summary2["Per Student Budget"])

summary2["%Passing Math"] = summary2["%Passing Math"].replace('%', '', regex=True).astype('float')
summary2["%Passing Math"] = pd.to_numeric(summary2["%Passing Math"])

summary2["%Passing reading"] = summary2["%Passing reading"].replace('%', '', regex=True).astype('float')
summary2["%Passing reading"] = pd.to_numeric(summary2["%Passing reading"])

# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

summary2['Spending Range Per Student'] = pd.cut(summary2['Per Student Budget'], bins = spending_bins, labels = group_names)

by_spending_range = summary2.groupby(['Spending Range Per Student'])

mean_math_score = by_spending_range["Average Math Score"].mean()
mean_reading_score = by_spending_range["Average Reading Score"].mean()
mean_percent_math = by_spending_range["%Passing Math"].mean()
mean_percent_reading = by_spending_range["%Passing reading"].mean()
overall_percent_pass = (mean_percent_math + mean_percent_reading)/2


scores_by_school_spending = pd.DataFrame({"Average Math Score":mean_math_score,
                                          "Average Reading Score":mean_reading_score,
                                          "%Passing Math":mean_percent_math,
                                          "%Passing reading":mean_percent_reading,
                                          "%Overall Passing Rate":overall_percent_pass
                                         })



scores_by_school_spending['%Passing Math'] = (scores_by_school_spending['%Passing Math']).apply(lambda x: '%{:,.2f}'.format(x))
scores_by_school_spending['%Passing reading'] = (scores_by_school_spending['%Passing reading']).apply(lambda x: '%{:,.2f}'.format(x))
scores_by_school_spending['%Overall Passing Rate'] = (scores_by_school_spending['%Overall Passing Rate']).apply(lambda x: '%{:,.2f}'.format(x))
scores_by_school_spending['Average Math Score'] = (scores_by_school_spending['Average Math Score']).apply(lambda x: '{:,.2f}'.format(x))
scores_by_school_spending['Average Reading Score'] = (scores_by_school_spending['Average Reading Score']).apply(lambda x: '{:,.2f}'.format(x))



scores_by_school_spending.head()
```
